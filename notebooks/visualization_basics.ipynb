```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/students.csv")
```

```python
print(df.head())
print(df.tail())
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(df["Marks"])
print(df[df["Marks"] > 88])
```

```python
plt.hist(df["Marks"])

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")

plt.show()
```

```python
plt.scatter(df["Age"], df["Marks"])

plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks")

plt.show()
```
