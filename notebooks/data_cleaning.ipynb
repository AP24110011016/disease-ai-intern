```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/students.csv")

# Display dataset
df
```

```python
print("First 5 Rows:")
print(df.head())
```

```python
print("Last 5 Rows:")
print(df.tail())
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Marks Column:")
print(df["Marks"])
```

```python
print("Name and Marks:")
print(df[["Name", "Marks"]])
```

```python
print("Students with Marks > 88:")
print(df[df["Marks"] > 88])
```

```python
plt.hist(df["Marks"])

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Marks Distribution")

plt.show()
```

```python
plt.scatter(df["Age"], df["Marks"])

plt.xlabel("Age")
plt.ylabel("Marks")
plt.title("Age vs Marks")

plt.show()
```
